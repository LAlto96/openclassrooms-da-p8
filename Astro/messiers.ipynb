```python
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
```

# Import Dataframe

```python

df_messier = pd.read_csv('./DonnéesAstro/catalogue-de-messier.csv', delimiter=';')
df_messier=df_messier.dropna()
df_messier.head(5)
```

# Nettoyage données

```python
def julian_to_datetime(julian_day):
    return datetime(2000, 1, 1) + timedelta(float(julian_day) - 2451545.0)
```

```python
# Function to convert Right Ascension value to degrees
def hms_to_deg(hms):
    "Converts 'h m s' to degrees"
    h, m, s = hms.split(':')
    return (float(h) + float(m) / 60 + float(s) / 3600)
```

```python
# Function to convert Declination value to degrees
def dms_to_decimal(dms):
    """Convert declination in DMS format to decimal format."""
    d, m, s = dms.split(':')
    return float(d) + float(m)/60 + float(s)/3600
```

```python
df_messier.columns
```

```python
df_messier['RA (Right Ascension)'] = df_messier['RA (Right Ascension)'].apply(hms_to_deg)
df_messier['Dec (Declinaison)'] = df_messier['Dec (Declinaison)'].apply(dms_to_decimal)
```
