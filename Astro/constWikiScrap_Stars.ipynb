# Librairies

```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
from time import sleep
```

# Import des liens

```python
df_link = pd.read_csv('contWikiScrap_Link.csv')
df_link[df_link['constellation']=='Mensa']
```

```python
url = df_link['url'].tolist()
len(url)
```

# Scrap des données sur les 88 constellations

```python
df_all = []
for i_url in range(len(url)-1):
    item_url = url[i_url]
    page = requests.get(item_url)
    soup = BeautifulSoup(page.text)

    rows = []
    table=soup.find_all('table')[0]
    
    column_titles=[]
    for tr in table.find_all('tr'):
        for th in tr.find_all('th'):
            column_titles.append(th.text.strip())
    columntitles = column_titles[:len(column_titles)-1]
    
    tablecontent=[]
    count=0
    for tbody in table.find_all('tbody'):
        for tr in tbody.find_all('tr'):
            count += 1
            for td in tr.find_all('td'):
                tablecontent.append(td.text)
    tbc=tablecontent[0:len(tablecontent)-6]
    
    table=[]
    for i in range(int(len(tbc)/len(columntitles)-1)):
        table.append(tbc[0:len(column_titles)-1])
        del tbc[0:len(column_titles)-1]
        
    df = pd.DataFrame(table, columns=columntitles)
    print(i)
    df['Constellation'] = df_link['constellation'][i_url]
    df = df[['Constellation', 'Name', 'RA', 'Dec', 'Dist. (ly)', 'vis.mag.', 'abs.mag.']]
    df_all.append(df)
    sleep(5)
    
```

```python
len(df_all)
```

## Mise dans un DataFrame des données et export CSV

```python
df_stars_by_constellation = pd.concat(df_all)
df_stars_by_constellation
```

```python
df_stars_by_constellation.to_csv('constWikiScrap_Stars.csv', index=False)
```
