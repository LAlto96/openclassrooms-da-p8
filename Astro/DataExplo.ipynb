```python
RA = df_messier['RA (Right Ascension)'].tolist()
RA[0]
```

```python
def hms_to_degrees(hms):
    h, m, s = hms.split(':')
    #print(h, m, s)
    return float(h) + float(m)/60 + float(s)/3600
```

```python
print(hms_to_degrees(RA[0]))
```

```python
RA_deg = [hms_to_degrees(hms) for hms in RA]
RA_deg[0]
```
